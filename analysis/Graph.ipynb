```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline # 그래프를 브라우저에 나타내기 위한 코드

# 데이터를 불러온다.
df = pd.read_csv("mtcar.csv")
df.plot()
plt.show()
```

```python
# 한개의 변수만 나타내보기: hp

df2 = df.hp #df.hp.plot()과 같은말
df2.plot()
```

```python
# 선 그래프: 여러개의 변수 그래프로 나타내기

df2 = df[['hp','disp']]
df2.plot()
```

```python
# 위의 그래프를 막대로 나타내보기
# kind에 그래프 종류를 전달해 준다

df2.plot(kind='bar')
```

```python
# bar그래프 다른 방법으로 그리기

df2.plot.bar()
```

```python
# pandas의 DataFrame객체 메서드가 아닌 matplotlib의 함수를 바로 실행해 그래프 그리기
plt.plot(np.arange(len(df.hp)), df.hp)
```

### 막대 그래프에 있는 몇 가지 option

```python
#1. 누적된 막대그래프
df2.plot.bar(stacked=True)
```

```python
#2. 가로로 누운 그래프 그리기(barh())
df2.plot.barh(stacked=True)
```

### 히스토그램 그리기

```python
# basic histograph

df1 = df.hp
df1.plot.hist()
```

```python
# 인자 조정: bins이용

df1.plot.hist(bins = 20)
```

```python
# 투명도 조정: alpha이용 _ 겹치는 그래프가 있을때 효과적으로 사용할 수 있다.

df2.plot.hist(alpha = 0.6)
```

### BoxPlot 그리기

- Box Plot은 중앙값, IQR, 아웃라이어 등 데이터의 특징을 모두 나타낸다.

```python
# Box plot을 나타내는 두 가지 방법

print('first')
print('첫번째',df2.boxplot())
print('second')
print('두번째',df2.plot.box())
```

```python
# 그룹 지어 그리기
#am: 자동기어 여부를 나타내는 변수로 0 혹은 1

df3 = df[['hp', 'disp', 'am']]
df3.boxplot(by = 'am')
```

### 산점도 그리기
- 두 변수 간의 상관관계를 보기위해 그린다. 

```python
# 기본적인 산점도

df3.plot.scatter(x='hp', y='disp')
```

```python
# 산점도의 원의 크기와 투명도 조절
# 원의 크기: s
# 투명도: alpha

df3.plot.scatter(x = 'hp', y = 'disp', s= 100, alpha=0.5)
```

```python
# pandas 패키지: DataFrame의 모든 변수에 대한 산점도를 한번에 보여줌

from pandas.plotting import scatter_matrix

scatter_matrix(df, alpha=0.4, figsize = (12, 12), diagonal='kde')
```
